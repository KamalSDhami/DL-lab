### Create a program to implement the working of LSTM – for the prediction of next number from the list of number provided.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Step 1: Define a number sequence
sequence = [i for i in range(1, 21)]  # Example: 1 to 20

# Step 2: Prepare input-output pairs
def create_sequences(seq, n_steps):
    X, y = [], []
    for i in range(len(seq) - n_steps):
        X.append(seq[i:i + n_steps])
        y.append(seq[i + n_steps])
    return np.array(X), np.array(y)

n_steps = 3  # How many previous numbers to use for prediction
X, y = create_sequences(sequence, n_steps)

# Step 3: Reshape X for LSTM [samples, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Step 4: Define LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Step 5: Train the model
model.fit(X, y, epochs=300, verbose=0)

# Step 6: Predict the next number
def predict_next_number(input_seq):
    input_seq = np.array(input_seq).reshape((1, n_steps, 1))
    return model.predict(input_seq, verbose=0)[0][0]

# Test prediction
test_input = [18, 19, 20]
predicted_number = predict_next_number(test_input)
print(f"Input: {test_input} → Predicted next number: {round(predicted_number, 2)}")
```
