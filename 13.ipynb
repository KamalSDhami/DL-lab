#### Create a program to implement the working of simple RNN - prediction of next letter of the word HELLO

```python
import tensorflow as tf
import numpy as np 
from tensorflow.keras.layers import SimpleRNN, Dense 
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
```

```python
text = "Hello"
char = sorted(set(text))
print(char)
char_to_index = {c:i for i,c in enumerate (char)}
index_to_text = {i:c for c,i in char_to_index.items()}
print(char_to_index)
print(index_to_text)
```

```python
input_char=text[:-1]
target_char = text[1:]
print(input_char)
print(target_char)
```

```python
input_index=[char_to_index[c] for c in input_char]
target_index = [char_to_index[c]for c in target_char]
print(input_index)
print(target_index)
```

```python
X = to_categorical(input_index,num_classes=len(char))
print(X)
print(X.shape)
X = np.reshape(X,(X.shape[0],1,X.shape[1]))
X
```

```python
y = to_categorical(target_index, num_classes=len(char))
```

```python
model = Sequential()
model.add(SimpleRNN(32, input_shape=(1, len(char)), return_sequences=False))
model.add(Dense(len(char), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

model.fit(X, y, epochs=500)
```

```python
chars = char
def predict_next_letter(char):
    input_idx = char_to_index[char]
    input_onehot = to_categorical([input_idx], num_classes=len(chars))
    input_onehot = np.reshape(input_onehot, (1, 1, len(chars)))
    prediction = model.predict(input_onehot,verbose=False)
    predicted_index = np.argmax(prediction)
    return index_to_text[predicted_index]

print("Predictions:")
for c in input_char:
    print(f"{c}  {predict_next_letter(c)}")
```
